```python
cars['Mileage']=cars['Mileage'].str.split().str[0]
cars['Mileage']
```

```python
cars['Mileage']=pd.to_numeric(cars['Mileage'])
```

```python
 cars['Mileage'].mean()
```

```python
cars['Mileage']=cars['Mileage'].fillna(round(cars['Mileage'].mean(),2))
cars['Mileage']
```

```python
cars['Seats']=cars['Seats'].fillna(round(cars['Seats'].mean(),2))
cars['Seats']
```

```python
cars['Engine']=cars['Engine'].str.split().str[0]
cars['Engine']
```

```python
cars['Engine']=pd.to_numeric(cars['Engine'])
```

```python
cars['Engine']=cars['Engine'].fillna(round(cars['Engine'].mean(),2))
cars['Engine']
```

```python
cars['Power']=cars['Power'].str.split().str[0]
cars['Power']
```

```python
cars['Power']=pd.to_numeric(cars['Power'])
```

```python
cars[cars['Power']=='null']
```

```python
cars['Power']=cars['Power'].replace('null',0)
cars['Power']
```

```python
cars['Power']=pd.to_numeric(cars['Power'])
```

```python
cars['Power']=cars['Power'].fillna(round(cars['Power'].mean(),2))
cars['Power']
```

```python
cars['Fuel_Type'].value_counts()
```

```python
ft_old=cars['Fuel_Type'].value_counts().index
ft_new=list(range(len(cars['Fuel_Type'].value_counts().index)))
print(ft_old)
print(ft_new)
```

```python
cars['Fuel_Type']=cars['Fuel_Type'].replace(ft_old,ft_new)
```

```python
loc_old=cars['Location'].value_counts().index
loc_new=list(range(len(cars['Location'].value_counts().index)))
cars['Location']=cars['Location'].replace(loc_old,loc_new)
```

```python
tra_old=cars['Transmission'].value_counts().index
tra_new=list(range(len(cars['Transmission'].value_counts().index)))
cars['Transmission']=cars['Transmission'].replace(tra_old,tra_new)
```

```python
ot_old=cars['Owner_Type'].value_counts().index
ot_new=list(range(len(cars['Owner_Type'].value_counts().index)))
cars['Owner_Type']=cars['Owner_Type'].replace(ot_old,ot_new)
```

```python
cars['Brands']=cars['Name'].str.split().str[0]
cars['Brands']
```

```python
cars['Model']=cars['Name'].str.split().str[1]
cars['Model']
cars=cars.drop('Name',axis=1)
```

```python
br_old=cars['Brands'].value_counts().index
br_new=list(range(len(cars['Brands'].value_counts().index)))
cars['Brands']=cars['Brands'].replace(br_old,br_new)
```
